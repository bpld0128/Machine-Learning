# California Housing Dataset Starter Code
The below code is a useful starting point that should get you started with exploring the California Housing data.

```python
import numpy as np
from sklearn.datasets.california_housing import fetch_california_housing
cal_housing = fetch_california_housing('.')
```

```python
print(cal_housing.DESCR)
```

```python
X = cal_housing.data
avg_price = cal_housing.target
feature_names = cal_housing.feature_names
feature_names
```

Some people might find it useful to use a [Pandas](http://synesthesiam.com/posts/an-introduction-to-pandas.html) [dataframe](http://www.gregreda.com/2013/10/26/intro-to-pandas-data-structures/) to [manipulate the data](http://www.gregreda.com/2013/10/26/working-with-pandas-dataframes/), but that is not really required for basic plotting and visualization: 

```python
import pandas as pd
df=pd.DataFrame(np.hstack([X,avg_price[:,np.newaxis]]),columns=np.hstack([feature_names,'price']))
#Example of the first few rows of data (using the "head" command)
df.head()
```

```python
# Below is how you might select only some columns from a dataframe.
# This is just an illustrative example; you could select other columns
subdf = df[['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','price']]
subdf.tail()
```

# Task 1: Visualize the relationship between features and price
Visualize the 2D relationship between housing prices and the provided features of the data. You can choose how you want to do this. 

[Enter your code into the empty cell below to create the necessary visualizations. You can create multiple cells for code or Markdown code if that helps you craft your submission.]


```python
# Code goes here
import matplotlib.pyplot as plt
import numpy as np

subdf = df[['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','price']]
x_MedInc = subdf['MedInc']
x_HouseAge = subdf['HouseAge']
x_AveRooms = subdf['AveRooms']
x_AveBedrms = subdf['AveBedrms']
x_Population = subdf['Population']
x_AveOccup = subdf['AveOccup']
y = subdf['price']

# MedInc = (np.mean(x_MedInc), np.std(x_MedInc))
# Price = (np.mean(y), np.std(y))
# print('Feature - MedInc: mean = %.3f stdv = %.3f' % (Feature))
# print('Price - Price : mean = %.3f stdv = %.3f' % (Price))

# # plot all in the same figure

# Creates subplots
f, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1, 6, sharey=True)
f.set_figheight(6)
f.set_figwidth(24)
ax1.scatter(x_MedInc, y, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax1.set_xlabel('MedInc')
ax2.scatter(x_HouseAge, y, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax2.set_xlabel('HouseAge')
ax3.scatter(x_AveRooms, y, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax3.set_xlabel('AveRooms')
ax4.scatter(x_AveBedrms, y, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax4.set_xlabel('AveBedrms')
ax5.scatter(x_Population, y, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax5.set_xlabel('Population')
ax6.scatter(x_AveOccup, y, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax6.set_xlabel('AveOccup')
ax1.set_ylabel('Price')
plt.show()
```

# (743 Only) Task 3: Visualize relationships w/ Anomalies Removed.
Using your knowledge of the anomalies you found above, remove those anomalies using appropriate code below (either by removing the entire data record, or just the specific values that were anomalous, if you prefer to be more surgical) and replot the Task 1 and Task 2 plots you produced above.

```python
# Code goes here for dealing with outliers
# You can copy/paste some of your plotting code from above, if that is helpful.

# Applying filter to remove strange anomalies
new_population = subdf['Population']<3000
new_HouseAge = subdf['HouseAge']<50
new_MedInc = subdf['MedInc']<8
new_AveOccup = subdf['AveOccup']<6
new_AveBedrms = subdf['AveBedrms']<2
new_AveRooms1 = subdf['AveRooms']<8
new_AveRooms2 = subdf['AveRooms']>2
new_price = subdf['price']<5

new_df = subdf[new_population & new_HouseAge & new_MedInc & new_AveOccup & new_AveBedrms & new_AveRooms1 & new_AveRooms2 & new_price]

# plot distribution plot
# sns.pairplot(new_df)
# sns.distplot(new_df['Population'])

# Obtain new data
xnew_MedInc = new_df['MedInc']
xnew_HouseAge = new_df['HouseAge']
xnew_AveRooms = new_df['AveRooms']
xnew_AveBedrms = new_df['AveBedrms']
xnew_Population = new_df['Population']
xnew_AveOccup = new_df['AveOccup']
ynew = new_df['price']

#compare the max of old and new values
# subdf.max()
# new_df.max()

# Plot the data with anomalies removed
f, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1, 6, sharey=True)
f.set_figheight(6)
f.set_figwidth(24)
ax1.scatter(xnew_MedInc, ynew, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax1.set_xlabel('MedInc')
ax2.scatter(xnew_HouseAge, ynew, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax2.set_xlabel('HouseAge')
ax3.scatter(xnew_AveRooms, ynew, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax3.set_xlabel('AveRooms')
ax4.scatter(xnew_AveBedrms, ynew, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax4.set_xlabel('AveBedrms')
ax5.scatter(xnew_Population, ynew, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax5.set_xlabel('Population')
ax6.scatter(xnew_AveOccup, ynew, facecolors = 'none', edgecolors = 'grey', alpha = '0.5')
ax6.set_xlabel('AveOccup')
ax1.set_ylabel('Price')
plt.show()
```
